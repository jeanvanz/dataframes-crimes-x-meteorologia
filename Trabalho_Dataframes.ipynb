```python
def estacao(date):
    month = date.month
    if month in [12, 1, 2]:
        return 'verao'
    elif month in [3, 4, 5]:
        return 'outono'
    elif month in [6, 7, 8]:
        return 'inverno'
    else:
        return 'primavera'

df_soledade_new['estacao'] = df_soledade_new['data_fato'].apply(estacao)
```

```python
def horario(hora_completa):
  hora, minuto, segundo = hora_completa.split(':')
  hora_int = int(hora)

  if hora_int >= 7 and hora_int < 13:
    return 'dia'
  elif hora_int >= 13 and hora_int < 19:
    return 'tarde'
  elif hora_int >= 19 and hora_int < 23:
    return 'noite'
  else:
    return 'madrugada'

df_soledade_new['horario'] = df_soledade_new['hora_fato'].apply(horario)
```

```python
crimes_por_horario = df_soledade_new.groupby(df_soledade_new['horario']).size()

crimes_por_horario.plot(kind='barh')
plt.title('Quantidade de Crimes por Horario')
plt.xlabel('Horario')
plt.ylabel('Crimes')
print(crimes_por_horario)
plt.show()
```
